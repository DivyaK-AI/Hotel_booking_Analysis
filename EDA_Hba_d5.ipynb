#Know Your Data

hotel : 1)Resort Hotel 2)City Hotel).

is_canceled :if the booking was canceled (1) or not canceled (0)

lead_time:Number of days that elapsed between the entering date of the booking into the PMS and the arrival date.

arrival_date_year : arriving Year date

arrival_date_month: arriving Month date

arrival_date_week_number : Week number in year for arrival date

arrival_date_day_of_month : Day of arrival date

stays_in_weekend_nights : Number of weekend nights (Saturday or Sunday)

stays_in_week_nights : Number of stay in week nights i.e Monday to Friday

adults : Number of adults

children : Number of children

babies: Number of babies

meal: Type of meal preferred.

Undefined/SC – no meal package

BB – Bed & Breakfast

HB – Half board (breakfast and one other meal)

FB – Full board (breakfast, lunch and dinner)

country : Country of origin

market_segment : Market segment designation.

TA:- Travel Agents and TO:-Tour Operators

distribution_channel: Booking distribution channel. The term “TA” means “Travel Agents” and “TO” means “Tour Operators”

is_repeated_guest : The booking name was from a repeated guest (1) or not (0)

previous_cancellations : Number of previous bookings that were cancelled by the customer compare to the current booking

previous_bookings_not_canceled : Number of previous bookings not cancelled by the customer compare to the current booking

reserved_room_type : Code of room type reserved.

assigned_room_type : Code for the type of room assigned to the booking.

booking_changes : Number of changes/amendments made to the booking from the moment the booking was entered on the PMS until the moment of check-in or cancellation

deposit_type: customer made a deposit to guarantee the booking. This variable can assume three categories:

No Deposit – no deposit was made

Non Refund - a deposit was made in the value of the total stay

cost Refundable – a deposit was made with a value under the total cost of stay.

agent : ID of the travel agency that made the booking

company : ID of the company/entity that made the booking or responsible for paying the booking. ID is presented instead of designation for anonymity reasons

days_in_waiting_list : Number of days the booking was in the waiting list before it was confirmed to the customer

customer_type : Type of booking, assuming one of four categories:

Contract - when the booking has an allotment or other type of contract associated to it

Group– when the booking is associated to a group

Transient– when the booking is not part of a group or contract, and is not associated to other transient booking

Transient-party – when the booking is transient, but is associated to at least other transient booking

adr: It is measured as the total revenues generated by all the occupied rooms in a hotel or lodge divided by the total number of occupied rooms over a given time period. It is a simple average that shows the revenues generated per occupied room.

The average daily rate (ADR) is a useful tool to maximize revenues in the hospitality sector. The ADR is measured as the total revenues generated by all the occupied rooms in the hotel or lodge divided by the total number of occupied rooms over a given time period. The average daily rate includes only the occupied rooms and not the total available stock. List item

required_car_parking_spaces : Number of car parking spaces required by the customer

*total_of_special_requests *: Number of special requests made by the customer

reservation_status : Reservation last status, assuming one of three categories:

Check-Out– customer has checked in but already departed

No-Show – customer did not check-in and did inform the hotel of the reason

reservation_status_date : Date at which the last status was set.

````python


# Define upper limits for outlier capping
lead_time_limit = 500
weekend_nights_limit = 5
adults_limit = 4
previous_bookings_limit = 1
previous_cancellations_limit = 1
week_nights_limit = 10
booking_changes_limit = 5
parking_spaces_limit = 5
children_limit = 8


# Cap outliers using np.clip function
df['lead_time'] = np.clip(df['lead_time'], None, lead_time_limit)
df['stays_in_weekend_nights'] = np.clip(df['stays_in_weekend_nights'], None, weekend_nights_limit)
df['adults'] = np.clip(df['adults'], None, adults_limit)
df['previous_bookings_not_canceled'] = np.clip(df['previous_bookings_not_canceled'], None, previous_bookings_limit)
df['previous_cancellations'] = np.clip(df['previous_cancellations'], None, previous_cancellations_limit)
df['stays_in_week_nights'] = np.clip(df['stays_in_week_nights'], None, week_nights_limit)
df['booking_changes'] = np.clip(df['booking_changes'], None, booking_changes_limit)
df['required_car_parking_spaces'] = np.clip(df['required_car_parking_spaces'], None, parking_spaces_limit)
df['children'] = np.clip(df['children'], None, children_limit)

````

````python
# Calculate revenue based on the specified conditions for Resort Hotel
revenue_0_resort = df[(df['hotel'] == 'Resort Hotel') & (df['is_canceled'] == 0) & ((df['deposit_type'] == 'No Deposit') | (df['deposit_type'] == 'Non Refund'))]['adr'].sum() - df[(df['hotel'] == 'Resort Hotel') & (df['is_canceled'] == 0) & (df['deposit_type'] == 'Refundable')]['adr'].sum()
revenue_1_resort = df[(df['hotel'] == 'Resort Hotel') & (df['is_canceled'] == 1) & (df['deposit_type'] == 'Non Refund')]['adr'].sum() - df[(df['hotel'] == 'Resort Hotel') & (df['is_canceled'] == 1) & (df['deposit_type'] == 'Refundable')]['adr'].sum()

# Calculate revenue based on the specified conditions for City Hotel
revenue_0_city = df[(df['hotel'] == 'City Hotel') & (df['is_canceled'] == 0) & ((df['deposit_type'] == 'No Deposit') | (df['deposit_type'] == 'Non Refund'))]['adr'].sum() - df[(df['hotel'] == 'City Hotel') & (df['is_canceled'] == 0) & (df['deposit_type'] == 'Refundable')]['adr'].sum()
revenue_1_city = df[(df['hotel'] == 'City Hotel') & (df['is_canceled'] == 1) & (df['deposit_type'] == 'Non Refund')]['adr'].sum() - df[(df['hotel'] == 'City Hotel') & (df['is_canceled'] == 1) & (df['deposit_type'] == 'Refundable')]['adr'].sum()

# Print the calculated revenues for Resort Hotel
#print("Revenue for Resort Hotel when not cancelled :", revenue_0_resort)
#print("Revenue for Resort Hotel when cancelled :", revenue_1_resort)

# Print the calculated revenues for City Hotel
#print("Revenue for City Hotel when not cancelled :", revenue_0_city)
#print("Revenue for City Hotel when cancelled : ", revenue_1_city)

# Calculate total revenue
total_revenue = revenue_0_resort + revenue_1_resort + revenue_0_city + revenue_1_city
print("Total Revenue:", total_revenue)
````

````python


# Define the order of months
months_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# Convert 'arrival_date_month' to a categorical variable with the specified order
df['Month'] = pd.Categorical(df['arrival_date_month'], categories=months_order, ordered=True)

# Group the data by 'Month' and 'arrival_date_year' and calculate the sum of 'total_people'
grouped_df = df.groupby(['Month', 'arrival_date_year'])['total_people'].sum().reset_index()

# Pivot the grouped data for plotting
pivot_df = grouped_df.pivot(index='Month', columns='arrival_date_year', values='total_people')

# Grouping by year and month to count cancellations
cancel_counts = df.groupby(['arrival_date_year', 'Month'])['is_canceled'].sum().reset_index()
# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(18, 6))

# Plot total people visited by month across years
sns.lineplot(data=pivot_df, x='Month', y=2015, marker='o', ax=axs[0], label='2015')
sns.lineplot(data=pivot_df, x='Month', y=2016, marker='o', ax=axs[0], label='2016')
sns.lineplot(data=pivot_df, x='Month', y=2017, marker='o', ax=axs[0], label='2017')
axs[0].set_title('Total People visited by Month Across Years (2015-2017)')
axs[0].set_xlabel('Month')
axs[0].set_ylabel('Total People')
axs[0].legend(title='Year', loc='upper left')
axs[0].tick_params(axis='x', rotation=45)

# Plot cancellation counts by year and month
sns.lineplot(data=cancel_counts, x='Month', y='is_canceled', hue='arrival_date_year', marker='o', ax=axs[1])
axs[1].set_title('Cancellation Counts by Month Across Years (2015-2017)')
axs[1].set_xlabel('Month')
axs[1].set_ylabel('Cancellation Count')
axs[1].legend(title='Year', loc='upper right')
axs[1].tick_params(axis='x', rotation=45)

fig = plt.gcf()
fig.patch.set_facecolor('#BCC6CC')

# Adjust layout
plt.tight_layout()
plt.show()
````

For the first plot, a line chart was chosen to visualize the trend in the total number of people visiting by month across the years 2015 to 2017. This type of chart is suitable for showing trends over time, which allows for easy comparison between different years.

For the second plot, a line chart was chosen to visualize the cancellation counts by month across the years 2015 to 2017. This chart type effectively shows the variation in cancellation counts over time and allows for comparison between different years.

2. What is/are the insight(s) found from the chart?

From the first plot, we can see the trend in the total number of people visiting each month across the years.we can clearly see that in august is the month people visited the most in each year.
From the second plot, we can see the cancellation counts for each month across the years. we can clearly see that in august is the month people cancelled the most in each year.
